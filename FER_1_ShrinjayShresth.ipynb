```python
#Import Necessary Libraries
!pip install deepface
import cv2
import matplotlib.pyplot as plt
from deepface import DeepFace
```

```python
# Load an image with a face
image_path = '/content/happyboy.jpg'
image = cv2.imread(image_path)
```

```python
# Load an image with a face
image_path = '/content/happyboy.jpg'  # Make sure this path is correct
image = cv2.imread(image_path)

# Check if the image loaded successfully
if image is None:
    print("Error: Could not load image. Please check the file path.")
else:
    # Convert the image from BGR (OpenCV default) to RGB
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Display the image
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```

```python
predictions = DeepFace.analyze(image)
```

```python
predictions
```

```python
# Access the first prediction
first_prediction = predictions[0]

# Print the dominant emotion of the first prediction
print(first_prediction['dominant_emotion'])  # Output: happy
```

```python
type(predictions)
```

```python
# Perform facial emotion recognition
result = DeepFace.analyze(img_path=image_path, actions=['emotion'])

# Access the first element of the list (assuming it's a single image analysis)
first_result = result[0]

# Print the results
print("Emotion: ", first_result['dominant_emotion'])
print("Details: ", first_result['emotion'])
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
gray= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
#print(faceCascade.empty())
faces= faceCascade.detectMultiScale(gray,1.1,4)

#Draw a rectangle around the faces
for(x, y, w, h)in faces:
  cv2.rectangle(image, (x,y),(x+w,y+h), (0, 255, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX

#Use putText() method for
#Inserting text on video
cv2.putText(image,
             first_prediction['dominant_emotion'],
            (0, 50),
            font, 1,
            (0, 0, 255),
            2,
            cv2.LINE_4) ;
```

```python
plt.imshow(cv2.cvtColor(image,cv2.COLOR_BGR2RGB))
```
